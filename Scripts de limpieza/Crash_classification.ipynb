```python
import pandas as pd
```

```python
ruta = r"C:\Users\ricar\Downloads\Traffic_Crashes_-_Crashes_20251125.csv"
df = pd.read_csv(ruta)
df.head(20) 
```

```python
# Limpiamos CRASH_RECORD_ID: Quitamos espacios en blancos y si hay IDs vacíos los hacemos Nan
df["CRASH_RECORD_ID"] = (
    df["CRASH_RECORD_ID"]
        .astype(str)
        .str.strip()
)
df.loc[
    df["CRASH_RECORD_ID"].isin(["", "nan", "None", "NULL"]),
    "CRASH_RECORD_ID"
] = None
# Vemos si hay valores nulos
df['CRASH_RECORD_ID'].isna().sum() # Notamos que no hay
# Eliminamos duplicados en IDs
df = df.drop_duplicates(subset=["CRASH_RECORD_ID"])
```

```python
df.columns.tolist()
```

```python
cols = [
    'CRASH_RECORD_ID',
    'FIRST_CRASH_TYPE',
    'CRASH_TYPE',
    'PRIM_CONTRIBUTORY_CAUSE',
    'SEC_CONTRIBUTORY_CAUSE',
    'DAMAGE',
    'HIT_AND_RUN_I'
]

crash_classification = df[cols].copy()
crash_classification.head()
```

```python
crash_classification.shape      # Muestra cuántas filas y columnas tiene
crash_classification.info()     # Da información sobre los tipos de datos y valores nulos 
```

```python
import numpy as np
# Columnas tipo texto (varchar en Postgres)
str_cols = crash_classification.select_dtypes(include='object').columns

# Quitar espacios al inicio/fin
for col in str_cols:
    crash_classification[col] = crash_classification[col].str.strip()

# Unificar códigos de "faltante" a NA
missing_tokens = ['', 'NaN', 'NAN', 'nan', 'NULL', 'Null', 'null']
crash_classification.replace(missing_tokens, pd.NA, inplace=True)
```

```python
cat_cols = [
    'FIRST_CRASH_TYPE',
    'CRASH_TYPE',
    'PRIM_CONTRIBUTORY_CAUSE',
    'SEC_CONTRIBUTORY_CAUSE',
    'DAMAGE'
]

for col in cat_cols:
    print(f'\n---- {col} ----')
    print('Valores distintos:', crash_classification[col].nunique(dropna=True))
    print(crash_classification[col].value_counts(dropna=True).head(20))
```

```python
col = crash_classification['HIT_AND_RUN_I'].astype('string') \
                                           .str.strip() \
                                           .str.upper()

crash_classification['HIT_AND_RUN_I'] = col.map({'Y': True, 'N': False})

# Tipo booleano con nulos (True/False/<NA>)
crash_classification['HIT_AND_RUN_I'] = crash_classification['HIT_AND_RUN_I'].astype('boolean')
```

```python
crash_classification
```

```python
# Quitar espacios
crash_classification['CRASH_RECORD_ID'] = crash_classification['CRASH_RECORD_ID'].str.strip()

# Ver cuántos faltan / duplicados antes 
print("IDs nulos:", crash_classification['CRASH_RECORD_ID'].isna().sum())
print("IDs duplicados:", crash_classification['CRASH_RECORD_ID'].duplicated().sum())
```

```python
# Quitar filas sin ID
crash_classification = crash_classification.dropna(subset=['CRASH_RECORD_ID'])

# Quitar filas con ID duplicado (nos quedamos con la primera aparición)
crash_classification = crash_classification.drop_duplicates(subset=['CRASH_RECORD_ID'])
```

```python
crash_classification = crash_classification.where(
    pd.notnull(crash_classification),
    None
)
```

```python
crash_classification
```

```python
# Ver si ya no existen NaN, NULL
missing_tokens = ['', 'NaN', 'NAN', 'nan', 'NULL', 'Null', 'null']

for col in crash_classification.columns:
    n_bad = crash_classification[col].isin(missing_tokens).sum()
    print(f'{col}: {n_bad} valores todavía como texto "NaN"/"NULL"/""')
```

```python
col = 'HIT_AND_RUN_I'

# Filas donde es faltante
mask = crash_classification[col].isna()

# Mira algunos valores para ver qué hay exactamente
print(crash_classification.loc[mask, col].head(10).tolist())
```

```python
# Pasarla a object
crash_classification['HIT_AND_RUN_I'] = crash_classification['HIT_AND_RUN_I'].astype('object')

# Volver a aplicar la conversión general a None
crash_classification = crash_classification.where(
    pd.notnull(crash_classification),
    None
)
```

```python
col = 'HIT_AND_RUN_I'

# Filas donde es faltante
mask = crash_classification[col].isna()

# Mira algunos valores para ver qué hay exactamente
print(crash_classification.loc[mask, col].head(10).tolist())
```
