```python
import pandas as pd
df = pd.read_csv('Vehicles_clean.csv')
df.head(5)
```

```python
df.columns
```

```python
df.shape
```

```python
vehicle = df[['VEHICLE_ID', 'CRASH_UNIT_ID', 'CRASH_RECORD_ID', 'CRASH_DATE', 'UNIT_NO','UNIT_TYPE', 'NUM_PASSENGERS', 'VEHICLE_YEAR', 'MAKE', 'MODEL', 'VEHICLE_TYPE']]
vehicle.head(5)
```

```python
vehicle.shape
```

```python
#VEHICLE_ID
vehicle['VEHICLE_ID'] = pd.to_numeric(vehicle['VEHICLE_ID'], errors='coerce')
vehicle = vehicle.dropna(subset=['VEHICLE_ID'])
vehicle['VEHICLE_ID'] = vehicle['VEHICLE_ID'].astype('Int64')
vehicle = vehicle.drop_duplicates(subset=['VEHICLE_ID'])

#CRASH_UNIT_ID limpio

#CRASH_RECORD_ID 
vehicle['CRASH_RECORD_ID'] = vehicle['CRASH_RECORD_ID'].str.strip()

#CRASH_DATE 
vehicle['CRASH_DATE'] = pd.to_datetime(vehicle['CRASH_DATE'], errors='coerce')
vehicle = vehicle[vehicle['CRASH_DATE'] <= pd.Timestamp.today()]

#UNIT_NO 
vehicle['UNIT_NO'] = vehicle['UNIT_NO'].astype(str).str.replace(',', '')
vehicle['UNIT_NO'] = vehicle['UNIT_NO'].astype(str).str.replace("'", '')
vehicle['UNIT_NO'] = pd.to_numeric(vehicle['UNIT_NO'], errors='coerce')
vehicle.loc[(vehicle['UNIT_NO'] < 0) | (vehicle['UNIT_NO'] > 50), 'UNIT_NO'] = pd.NA
vehicle['UNIT_NO'] = vehicle['UNIT_NO'].astype('Int64')

#UNIT_TYPE limpio

#NUM_PASSENGERS
vehicle['NUM_PASSENGERS'] = pd.to_numeric(vehicle['NUM_PASSENGERS'], errors='coerce')
vehicle.loc[(vehicle['NUM_PASSENGERS'] < 0) | (vehicle['NUM_PASSENGERS'] > 70), 'NUM_PASSENGERS'] = pd.NA
vehicle['NUM_PASSENGERS'] = vehicle['NUM_PASSENGERS'].astype('Int64')
vehicle['NUM_PASSENGERS'] = vehicle['NUM_PASSENGERS'].fillna(0)

#VEHICLE_YEAR
vehicle['VEHICLE_YEAR'] = pd.to_numeric(vehicle['VEHICLE_YEAR'], errors='coerce')
vehicle.loc[(vehicle['VEHICLE_YEAR'] < 1900) | (vehicle['VEHICLE_YEAR'] > 2026), 'VEHICLE_YEAR'] = pd.NA
vehicle['VEHICLE_YEAR'] = vehicle['VEHICLE_YEAR'].astype('Int64')

missing_tokens = {
    '', ' ', 'UNKNOWN', 'UNKNOWN/NA', 'UNK',
    'N/A', 'NA', 'NONE', 'NULL', 'NAN',
    'OTHER', 'OTHER (EXPLAIN IN NARRATIVE)',
    'NOT APPLICABLE'
}
for col in ['MAKE', 'MODEL', 'VEHICLE_TYPE']:
    # 1. Use pandas StringDtype (handles <NA> nicely)
    vehicle[col] = vehicle[col].astype('string')
    vehicle[col] = vehicle[col].str.upper()
    vehicle.loc[vehicle[col].isin(missing_tokens), col] = pd.NA

```

```python
for col in vehicle.columns:
    print(f"\n--- {col} ---")
    print("Valores únicos:", vehicle[col].unique()[:50])  # muestra los primeros 10 valores únicos
    print("Cantidad de nulos:", vehicle[col].isna().sum())
    print("Tipo de dato:", vehicle[col].dtype)
```

```python
vehicle_violations = df[['VEHICLE_ID', 'CMRC_VEH_I', 'EXCEED_SPEED_LIMIT_I', 'HAZMAT_PRESENT_I', 'VEHICLE_DEFECT']]
vehicle_violations.head(5)
```

```python
#VEHICLE_ID
vehicle_violations['VEHICLE_ID'] = pd.to_numeric(vehicle_violations['VEHICLE_ID'], errors='coerce')
vehicle_violations = vehicle_violations.dropna(subset=['VEHICLE_ID'])
vehicle_violations['VEHICLE_ID'] = vehicle_violations['VEHICLE_ID'].astype('Int64')
vehicle_violations = vehicle_violations.drop_duplicates(subset=['VEHICLE_ID'])

boolean_map = {'Y': True, 'N': False}

# CMRC
vehicle_violations['CMRC_VEH_I'] = (
    vehicle_violations['CMRC_VEH_I']
    .map(boolean_map)
    .astype('boolean')
)

# EXCEED SPEED LIMIT
vehicle_violations['EXCEED_SPEED_LIMIT_I'] = (
    vehicle_violations['EXCEED_SPEED_LIMIT_I']
    .map(boolean_map)
    .astype('boolean')
)

# HAZMAT
vehicle_violations['HAZMAT_PRESENT_I'] = (
    vehicle_violations['HAZMAT_PRESENT_I']
    .replace('U', pd.NA)
    .map(boolean_map)
    .astype('boolean')
)
```

```python
for col in vehicle_violations.columns:
    print(f"\n--- {col} ---")
    print("Valores únicos:", vehicle_violations[col].unique()[:30])  # muestra los primeros 10 valores únicos
    print("Cantidad de nulos:", vehicle_violations[col].isna().sum())
    print("Tipo de dato:", vehicle_violations[col].dtype)
```
