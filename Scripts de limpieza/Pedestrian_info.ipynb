```python
import numpy as np
import pandas as pd

# Columnas tipo texto
str_cols = pedestrian_info.select_dtypes(include='object').columns

# Quitar espacios al inicio/fin
for col in str_cols:
    pedestrian_info[col] = pedestrian_info[col].str.strip()

# Unificar códigos de "faltante" a NA
missing_tokens = ['', 'NaN', 'NAN', 'nan', 'NULL', 'Null', 'null']
pedestrian_info.replace(missing_tokens, pd.NA, inplace=True)
```

```python
import pandas as pd

# Detectar columnas de texto (varchar)
str_cols = pedestrian_info.select_dtypes(include='object').columns

# Normalizar texto y reemplazar NAN, NA, UNKNOWN, UNKNOWN/NA → None
for col in str_cols:
    # Pasamos a string, quitamos espacios y ponemos en mayúsculas
    s = pedestrian_info[col].astype('string').str.strip().str.upper()
    
    # Reemplazamos los tokens que quieres tratar como faltantes
    s = s.replace({
        'NAN': None,
        'NA': None,
        'UNKNOWN': None,
        'UNKNOWN/NA': None
    })
    
    pedestrian_info[col] = s


# Convierte cualquier NaN / <NA> que quede a None en toda la tabla
pedestrian_info = pedestrian_info.where(
    pd.notnull(pedestrian_info),
    None
)
```
