```python
import pandas as pd
```

```python
# Esta es la entidad de Crashes
df = pd.read_csv('/Users/regina/Desktop/ITAM 3 sem/Bases de datos/Proyecto/Traffic_Crashes_-_Crashes_20251030.csv')
```

```python
# Definimos las columnas que eliminaremos
cols_to_drop = ['CRASH_DATE_EST_I', 'ALIGNMENT', 'REPORT_TYPE', 'DATE_POLICE_NOTIFIED', 'STREET_DIRECTION', 'LOCATION', 'INJURIES_UNKNOWN', 'INJURIES_NO_INDICATION', 'INJURIES_REPORTED_NOT_EVIDENT', 'INJURIES_NON_INCAPACITATING', 'MOST_SEVERE_INJURY', 'WORKERS_PRESENT_I', 
                'WORK_ZONE_TYPE', 'PHOTOS_TAKEN_I', 'DOORING_I', 'STATEMENTS_TAKEN_I', 'BEAT_OF_OCCURRENCE']
```

```python
df = df.drop(cols_to_drop, axis = 1)
```

```python
df.to_csv('Traffic_Crashes_clean.csv', index=False)
```

```python
# Normalizamos los nombres de las columnas
df.columns = (
    df.columns
      .astype(str)
      .str.replace("\ufeff", "", regex=False)       
      .str.replace(r"^[^A-Za-z0-9]+", "", regex=True) 
      .str.strip()
      .str.replace(r"\s+", "_", regex=True)         
)
```

```python
# Limpiamos CRASH_RECORD_ID: Quitamos espacios en blancos y si hay IDs vacíos los hacemos Nan
df["CRASH_RECORD_ID"] = (
    df["CRASH_RECORD_ID"]
        .astype(str)
        .str.strip()
)
df.loc[
    df["CRASH_RECORD_ID"].isin(["", "nan", "None", "NULL"]),
    "CRASH_RECORD_ID"
] = None
# Vemos si hay valores nulos
df['CRASH_RECORD_ID'].isna().sum() # Notamos que no hay
# Eliminamos duplicados en IDs
df = df.drop_duplicates(subset=["CRASH_RECORD_ID"])
```

```python
# Creamos la tabla Crash_Date
df_CRASH_DATE = df[['CRASH_RECORD_ID', 'CRASH_DAY_OF_WEEK', 'CRASH_MONTH']]
```

```python
# Limpieza de Crash_Date
# Confirmamos que todos los valores de day of week esten entre 1 y 7
df_CRASH_DATE['CRASH_DAY_OF_WEEK'].unique()
```

```python
# Confirmamos que todos los valores de month esten entre entre 1 y 12
df_CRASH_DATE['CRASH_MONTH'].unique()
```

```python
# Convertimos month y day of week a INT
df_CRASH_DATE["CRASH_DAY_OF_WEEK"] = df_CRASH_DATE["CRASH_DAY_OF_WEEK"].astype("Int64")
df_CRASH_DATE["CRASH_MONTH"] = df_CRASH_DATE["CRASH_MONTH"].astype("Int64")
```

```python
df_CRASH_DATE.info
```

```python
df_CRASH_DATE.to_csv('Crashes_Date.csv', index=False)
```

```python
# Creamos la tabla Crashes
df_CRASHES = df[['CRASH_RECORD_ID', 'CRASH_DATE', 'LATITUDE', 'LONGITUDE', 'STREET_NO', 'STREET_NAME']]
```

```python
# Limpieza de Crashes
# Convertimos CRASH_DATE a timestamp
df_CRASHES["CRASH_DATE"] = pd.to_datetime(df_CRASHES["CRASH_DATE"], errors="coerce")
```

```python
# Convertir latitude y longitude a numeric 
df_CRASHES["LATITUDE"] = pd.to_numeric(df_CRASHES["LATITUDE"], errors="coerce")
df_CRASHES["LONGITUDE"] = pd.to_numeric(df_CRASHES["LONGITUDE"], errors="coerce")
```

```python
# Limpiar Street_No y Street_Name
df_CRASHES["STREET_NO"] = (
    df_CRASHES["STREET_NO"]
      .astype(str)
      .str.strip()
)
df_CRASHES["STREET_NAME"] = (
    df_CRASHES["STREET_NAME"]
      .astype(str)
      .str.strip()
      .str.replace(r"\s+", " ", regex=True)
      .str.upper()
)
```

```python
# Si hay valores vacíos los volvemos Nan
df.loc[df["CRASH_DATE"].isin(["", "nan", "None", "NULL"]), "CRASH_DATE"] = None
df.loc[df["LATITUDE"].isin(["", "nan", "None", "NULL"]), "LATITUDE"] = None
df.loc[df["LONGITUDE"].isin(["", "nan", "None", "NULL"]), "LONGITUDE"] = None
df.loc[df["STREET_NO"].isin(["", "nan", "None", "NULL"]), "STREET_NO"] = None
df.loc[df["STREET_NAME"].isin(["", "nan", "None", "NULL"]), "STREET_NAME"] = None
# Checamos si hay nulos
df_CRASHES.isna().sum()
# Como son pocos, los dejamos 
```

```python
df_CRASHES.to_csv("Crashes.csv", index=False)
```

```python
# Creamos la tabla Crash_Circumstances
df_CIRCUMSTANCES = df[['CRASH_RECORD_ID', 'TRAFFIC_CONTROL_DEVICE', 'DEVICE_CONDITION', 'WEATHER_CONDITION', 'LIGHTING_CONDITION', 'LANE_CNT', 'ROADWAY_SURFACE_COND', 'ROAD_DEFECT', 'WORK_ZONE_I', 'NUM_UNITS', 'POSTED_SPEED_LIMIT', 'INTERSECTION_RELATED_I', 'NOT_RIGHT_OF_WAY_I']]
```

```python
# Limpieza en los tipo INT Lane_CNT, Num_Units, Posted_Speed_Limit 
df_CIRCUMSTANCES['POSTED_SPEED_LIMIT'].unique()
```

```python
df_CIRCUMSTANCES['LANE_CNT'].unique()
```

```python
df_CIRCUMSTANCES['NUM_UNITS'].unique()
```

```python
# Eliminamos los valores extraños en LANE_CNT
df_CIRCUMSTANCES["LANE_CNT"] = (
df_CIRCUMSTANCES["LANE_CNT"]
      .astype(str)
      .str.replace(",", "", regex=False)      
      .str.strip()
)

df_CIRCUMSTANCES["LANE_CNT"] = pd.to_numeric(
    df_CIRCUMSTANCES["LANE_CNT"], errors="coerce"
)

df_CIRCUMSTANCES.loc[~df_CIRCUMSTANCES["LANE_CNT"].between(0, 26), "LANE_CNT"] = None

df_CIRCUMSTANCES["LANE_CNT"] = df_CIRCUMSTANCES["LANE_CNT"].astype("Int64")
```

```python
# Limpieza en los tipo VARCHAR 
# Convertimos WORK_ZONE_I, INTERSECTION_RELATED_I, NOT_RIGHT_OF_WAY_I a boolean, pues sus únicos valores son Nan, Y, N
df_CIRCUMSTANCES["WORK_ZONE_I"] = (
    df_CIRCUMSTANCES["WORK_ZONE_I"]
        .astype(str).str.strip().str.upper()
        .replace({"Y": True, "N": False, "NAN": pd.NA, "": pd.NA, "NONE": pd.NA})
        .astype("boolean")
)
df_CIRCUMSTANCES["INTERSECTION_RELATED_I"] = (
    df_CIRCUMSTANCES["INTERSECTION_RELATED_I"]
        .astype(str).str.strip().str.upper()
        .replace({"Y": True, "N": False, "NAN": pd.NA, "": pd.NA, "NONE": pd.NA})
        .astype("boolean")
)
df_CIRCUMSTANCES["NOT_RIGHT_OF_WAY_I"] = (
    df_CIRCUMSTANCES["NOT_RIGHT_OF_WAY_I"]
        .astype(str).str.strip().str.upper()
        .replace({"Y": True, "N": False, "NAN": pd.NA, "": pd.NA, "NONE": pd.NA})
        .astype("boolean")
)
```

```python
# Limpiamos las categóricas
df_CIRCUMSTANCES["TRAFFIC_CONTROL_DEVICE"] = (
    df_CIRCUMSTANCES["TRAFFIC_CONTROL_DEVICE"]
      .astype(str).str.strip()
      .str.replace(r"\s+", " ", regex=True)
      .str.upper()
      .replace({"": None, "NAN": None, "NONE": None})
)
df_CIRCUMSTANCES["DEVICE_CONDITION"] = (
    df_CIRCUMSTANCES["DEVICE_CONDITION"]
      .astype(str).str.strip()
      .str.replace(r"\s+", " ", regex=True)
      .str.upper()
      .replace({"": None, "NAN": None, "NONE": None})
)
df_CIRCUMSTANCES["WEATHER_CONDITION"] = (
    df_CIRCUMSTANCES["WEATHER_CONDITION"]
      .astype(str).str.strip()
      .str.replace(r"\s+", " ", regex=True)
      .str.upper()
      .replace({"": None, "NAN": None, "NONE": None})
)
df_CIRCUMSTANCES["LIGHTING_CONDITION"] = (
    df_CIRCUMSTANCES["LIGHTING_CONDITION"]
      .astype(str).str.strip()
      .str.replace(r"\s+", " ", regex=True)
      .str.upper()
      .replace({"": None, "NAN": None, "NONE": None})
)
df_CIRCUMSTANCES["ROADWAY_SURFACE_COND"] = (
    df_CIRCUMSTANCES["ROADWAY_SURFACE_COND"]
      .astype(str).str.strip()
      .str.replace(r"\s+", " ", regex=True)
      .str.upper()
      .replace({"": None, "NAN": None, "NONE": None})
)
df_CIRCUMSTANCES["ROAD_DEFECT"] = (
    df_CIRCUMSTANCES["ROAD_DEFECT"]
      .astype(str).str.strip()
      .str.replace(r"\s+", " ", regex=True)
      .str.upper()
      .replace({"": None, "NAN": None, "NONE": None})
)
```

```python
# Vemos cuantos son nulos (en porcentaje)
(df_CIRCUMSTANCES.isna().mean() * 100).round(2)
```

```python
# Notando el alto porcentaje de nulos en WORK_ZONE_I, decidimos eliminarla
df_CIRCUMSTANCES = df_CIRCUMSTANCES.drop(columns=["WORK_ZONE_I"])
```

```python
df_CIRCUMSTANCES.to_csv("Crashes_Circumstances.csv", index=False)
```
