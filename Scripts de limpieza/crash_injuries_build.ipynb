```python
import pandas as pd
df = pd.read_csv('crashes_clean.csv')
```

```python
crash_injuries = df[['CRASH_RECORD_ID', 'INJURIES_FATAL', 'INJURIES_INCAPACITATING', 'INJURIES_TOTAL']].copy()
for col in crash_injuries.columns:
    print(f"\n--- {col} ---")
    print("Valores únicos:", crash_injuries[col].unique()[:30])  # muestra los primeros 30 valores únicos
    print("Cantidad de nulos:", crash_injuries[col].isna().sum())
    print("Tipo de dato:", crash_injuries[col].dtype)
```

```python
crash_injuries['INJURIES_OTHERS'] = (
    crash_injuries['INJURIES_TOTAL'] 
    - crash_injuries['INJURIES_FATAL'] 
    - crash_injuries['INJURIES_INCAPACITATING']
)
```

```python
crash_injuries = crash_injuries.drop(columns=['INJURIES_TOTAL'])
crash_injuries.head(5)
```

```python
crash_injuries['CRASH_RECORD_ID'] = (
    crash_injuries['CRASH_RECORD_ID']
    .astype('string')
    .str.strip()
    .str.lower()
)
crash_injuries = crash_injuries.drop_duplicates(subset=['CRASH_RECORD_ID'])

injury_cols = ['INJURIES_FATAL', 'INJURIES_INCAPACITATING', 'INJURIES_OTHERS']

for col in injury_cols:
    crash_injuries[col] = pd.to_numeric(crash_injuries[col], errors='coerce')
    crash_injuries.loc[crash_injuries[col] < 0, col] = pd.NA
    #crash_injuries.loc[crash_injuries[col] > 99, col] = pd.NA
    crash_injuries[col] = crash_injuries[col].astype('Int64')

#Para droppear filas con todas las lesiones nulas
# mask_all_null = crash_injuries[injury_cols].isna().all(axis=1)
# crash_injuries = crash_injuries[~mask_all_null]
```

```python
for col in crash_injuries.columns:
    print(f"\n--- {col} ---")
    print("Valores únicos:", crash_injuries[col].unique()[:30])  # muestra los primeros 30 valores únicos
    print("Cantidad de nulos:", crash_injuries[col].isna().sum())
    print("Tipo de dato:", crash_injuries[col].dtype)
```

```python
crash_injuries.to_csv('crash_injuries.csv', index=False)
```
